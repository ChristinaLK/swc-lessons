### Defining Defaults

```python
def norm_value(val, low, high):
    assert low < high, 'input not correct'
    return (val - low)/(high - low)

def normalize(data_array,lower=0.0,upper=1.0):
    minimum = data_array.min()
    maximum = data_array.max()
    for values in data_array:
        values = norm_value(values,minimum,maximum)
    
```

    and `False` otherwise.
    Compare your implementation with your partner's:
    do you return the same answer for all possible pairs of numbers?

```python
def near(p1, p2):
    percent = p2*.10
    safe_zone = (p2-percent, p2+percent)
    if p1 >= safe_zone[0] and p1<= safe_zone[1]:
        return True
    else:
        return False
```

```python
near(8,10)
```
