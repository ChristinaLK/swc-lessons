This function selects all entries in a range of years, and then counts...

```python
def year_range(df, lower, upper):
    assert type(lower) is int, "Incorrect type"
    assert type(upper) is int, "Incorrect type"
    assert lower < upper, "range is incorrect"
    return df.loc[((df.year > lower) & (df.year < upper)),["continent","pop"]].groupby("continent").sum().sort("pop")

print year_range(data, 1950, 1955)
```

This function creates a horizontal bar chart of the data created in the previous function

```python
def plot_vals(df, lower, upper):
    lower_str = str(lower)
    upper_str = str(upper)
    
    types = list(df.index)
    y_pos = np.arange(len(types))

    plt.barh(y_pos, df["pop"], align='center')
    plt.yticks(y_pos, types)
    plt.xlabel('Population')
    plt.ylabel('Continent')
    plt.title('Population from '+lower_str+' to '+upper_str)
    plt.show()

vals(year_range(data,1980,2000), 1980, 2000)
```
