```python
selected_data = df.loc[(df.Year > 2000),["Type","Key"]].groupby("Type").count().sort("Key")
```

```python
print selected_data
```

```python
types = list(selected_data.index)

y_pos = np.arange(len(types))

print y_pos
```

```python
import matplotlib.pyplot as plt
```

```python

plt.barh(y_pos, selected_data["Key"], align='center')
plt.yticks(y_pos, types)
plt.xlabel('Number of Publications')
plt.ylabel('Type of Publication')
plt.title('Number of Publications by Type')
```

```python
def plot_pubs(df):
    types = list(df.index)
    y_pos = np.arange(len(types))

    plt.barh(y_pos, df["Key"], align='center')
    plt.yticks(y_pos, types)
    plt.xlabel('Number of Publications')
    plt.ylabel('Type of Publication')
    plt.title('Number of Publications by Type')
    plt.show()
```

```python
plot_pubs(selected_data)
```
