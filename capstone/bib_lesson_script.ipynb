I'll be using the `pandas` library to read in my `.csv` file.  It has many more columns, but I'm only interested
in the first few.  

```python
data = pd.read_csv("data/bibliography.csv", header=0, usecols=range(6))
```

This function selects all publications in a range of years, and then counts the number of publications during that time frame by type. 

```python
def year_range(df, lower, upper):
    assert type(lower) is int, "Incorrect type"
    assert type(upper) is int, "Incorrect type"
    assert lower < upper, "range is incorrect"
    return df.loc[((df.Year > lower) & (df.Year < upper)),["Type","Key"]].groupby("Type").count().sort("Key")

print year_range(data, 1980, 2000)
```

This function creates a horizontal bar chart of the data created in the previous function

```python
def plot_pubs(df, lower, upper):
    lower_str = str(lower)
    upper_str = str(upper)
    
    types = list(df.index)
    y_pos = np.arange(len(types))

    plt.barh(y_pos, df["Key"], align='center')
    plt.yticks(y_pos, types)
    plt.xlabel('Number of Publications')
    plt.ylabel('Type of Publication')
    plt.title('Number of Publications From '+lower_str+' to '+upper_str)
    plt.show()

plot_pubs(year_range(data,1980,2000), 1980, 2000)
```

Now loop over years, producing graph for each block of 20

```python
for year in years:
    lower = year
    upper = year + 20
    plot_pubs(year_range(data,lower,upper), lower, upper)
```
